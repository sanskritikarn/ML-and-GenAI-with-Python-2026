```python
#Q2 Create a function to calculate sum of first N natural numbers.

num = int(input("Enter N: "))
def sum_natural_numbers(n):
    sum = 0
    for i in range(1,n+1):
       sum = sum+i
    print(sum)
sum_natural_numbers(num)
```

```python
#Q3 Create a function to reverse a number.

num = int(input("Enter the number: "))
def reverse_a_number(n):
    rev = 0
    while n>0:
        rem = n%10
        rev = rev*10 + rem
        n = n//10
    print("Reverse of a number is ",rev)
reverse_a_number(num)
```

```python
#Q4 Create a function to count digits in a number.

num = int(input("Enter the number: "))
def count_digits(n):
    sum = 0
    while n>0:
        sum = sum+1
        n = n//10
    print("No. of digits in a number are: ",sum)
count_digits(num)
```

```python
#Q5 Create a function to check palindrome number.

num = int(input("Enter the number: "))
def check_palindrome(n):
    temp = n
    rev = 0
    while n>0:
        rem = n%10
        rev = rev*10 + rem
        n = n//10
    if temp == rev:
        print(temp," is palindrome number")
    else:
        print(temp," is not a palindrome number")
check_palindrome(num)
```

```python
#Q7 Calculator Using Functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

choice = int(input("Enter choice: "))

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if choice == 1:
    print("Result =", add(a, b))
elif choice == 2:
    print("Result =", subtract(a, b))
elif choice == 3:
    print("Result =", multiply(a, b))
elif choice == 4:
    print("Result =", divide(a, b))
else:
    print("Invalid Choice")
```
