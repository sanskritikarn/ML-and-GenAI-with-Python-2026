```python
#Find area of rectangle. 
# Taking input from user
l = float(input("Enter length: "))
b = float(input("Enter breadth: "))
# Calculating area
area = l * b
# Display result
print("Area of rectangle is:", area)
```

```python
#Find simple interest
# Taking input
P = float(input("Enter principal amount: "))
R = float(input("Enter rate of interest: "))
T = float(input("Enter time (in years): "))
# Calculating Simple Interest
SI = (P * R * T) / 100
# Display result
print("Simple Interest is:", SI)
```

```python
#Convert temperature from Celsius to Fahrenheit
# Taking input
celsius = float(input("Enter temperature in Celsius: "))
# Converting to Fahrenheit
fahrenheit = (9/5) * celsius + 32
# Display result
print("Temperature in Fahrenheit is:", fahrenheit)
```

```python
#Calculate average of 3 numbers
# Taking input
a = int(input("Enter 1st number: "))
b = int(input("Enter 2nd number: "))
c = int(input("Enter 3rd number: "))
# Taking average of 3 numbers
avg = (a+b+c)/3
# Display result
print("Average of 3 numbers are:", avg)
```

```python
#Swap 2 numbers without third variable 
# Taking input
x = int(input("Enter 1st number"))
y = int(input("Enter 2nd number"))
# Swapping without third variable
x, y = y, x
# Display result
print("After swapping:")
print("First number =", x)
print("Second number=", y)
```

```python
#Create a Student Report Program that take student details using input(), Store marks in variables, 
#Calculate total and percentage , Use comments, Use proper indentation

# Student Report Program

# Taking student details as input
name = input("Enter the name: ")
roll_no = int(input("Enter the roll no.: "))
#Storing marks in variables
marks1 = int(input("Enter the marks of subject 1: "))
marks2 = int(input("Enter the marks of subject 2: "))
marks3 = int(input("Enter the marks of subject 3: "))
#Calculating total marks
total = marks1+marks2+marks3
#Calculating percentage
percentage = total/3
# Displaying student report
print("----- STUDENT REPORT -----")
print("Name:", name)
print("Roll No:", roll_no)
print("Subject 1 Marks:", marks1)
print("Subject 2 Marks:", marks2)
print("Subject 3 Marks:", marks3)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```
